# Matplotlib Tutorial Files 
###  Topics - pyplot.plt , format_string and other options

# Importing needed ibraries

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Data For Graph


```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

### Basic graph details

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
```

### Graph Potting Fuctions


```python
plt.plot(x,y)
```

### plotting the same graph with added features


```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, '--')
```

### More Graph Plotting Functions

```python
plt.plot(x,y,color='blue',marker='+',linestyle=' ')
```

### Playing with marker sizes


```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)
```

```python
plt.plot(x,y,'c+',linestyle = 'dotted', markersize = 20)
```

### Alpha property to control transparency of line chart

```python
plt.plot(x,y,'m<',alpha=0.3)
```
